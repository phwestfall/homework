```python
# import dependencies
import pandas as pd
import numpy as np
```

```python
# read in death data from wonder cdc as a csv that is tab delimited

overalldeath = pd.read_csv('Resources/OverallDeaths.txt', sep="\t")
```

```python
# rename columns, drop unnecessary rows, create new column calculating the total death rate per 1000 people

overalldeath = overalldeath.rename(columns={'Deaths': 'Total_Deaths'})
overalldeath = overalldeath.drop(['Crude Rate', 'Notes'], axis=1)
overalldeath["Total_Death_Rate"] = (overalldeath['Total_Deaths'] / overalldeath['Population']) * 1000
```

```python
# read in death by accident txt file, rename columns, calculate death rate, assign columns to dataframe
# merge with previous data set on County Code

accident = pd.read_csv('Resources/AccidentDeaths.txt', sep="\t")
accident = accident.rename(columns={'Deaths': 'Accident_Deaths'})
accident["Accident_Death_Rate"] = (accident['Accident_Deaths'] / accident['Population']) * 1000
accident = accident[["County Code", "Accident_Deaths", "Accident_Death_Rate"]]
merge1 = overalldeath.merge(accident, how="outer", left_on='County Code', right_on='County Code')
```

```python
# read ub svi csv from CDC's website
svi = pd.read_csv('Resources/SVI2016_US_COUNTY.csv')

# merge social vulernability index csv by FIPS county code with deathrate csv County Code. 
svi_merged = pd.merge(svi, deathrate, how="outer", on="FIPS") 
```

```python
# removes margin of error columns from the dataframe and flag 
svi_merged = svi_merged[svi_merged.columns.drop(list(svi_merged.filter(regex="M_")))]
svi_merged = svi_merged[svi_merged.columns.drop(list(svi_merged.filter(regex="F_")))]
```

```python
svi_merged.head()
```

```python
########### LIFE EXPECTANCY DATA ###################

# read in life expetancy data, remove column, drop index

life = pd.read_csv('Resources/lifeexpectancy.csv')
life.columns = life.iloc[0]
life = life.reindex(life.index.drop(0))

# clean up columns

life = life.rename(columns={'Life expectancy, 2014*': 'Life Expectancy'})
life = life[['Location', 'FIPS', 'Life Expectancy']]
life.head()

# take first five digits from life expectancy column

life['Life Expectancy'] = life['Life Expectancy'].str[:5]

# change FIPS to integer for merge

life['FIPS'] = pd.to_numeric(life['FIPS'], errors='coerce')
life = life.dropna(subset=['FIPS'])
life['FIPS'] = life['FIPS'].astype(int)

# remove columns that just have state overall data

life = life[life.FIPS > 100]
life = life.reset_index(drop=True)
```

```python
#################### MEDICARE COST DATA #####################

medicare = pd.read_csv('Resources/medicarecost.csv')



# clean up medicare dataframe by renaming and selectin columns

medicare = medicare.rename(columns={'Total Medicare reimbursements per enrollee (Parts A and B) (2016)':
                                    'Total per enrollee (age, sex, race adjusted)', 
                                    'Unnamed: 4': 'Total per enrollee (age, sex, race, price adjusted)'})

medicare = medicare[['County ID', 'County name', 'Total per enrollee (age, sex, race adjusted)',
                     'Total per enrollee (age, sex, race, price adjusted)']]

medicare = medicare.drop(medicare.index[0])

# make county id an int and rename it for merge purposes

medicare['County ID'] = medicare['County ID'].astype(int)
medicare = medicare.rename(columns={'County ID': 'FIPS'})

# merge life expectancy and medicare cost data together

lifemedicare = pd.merge(life, medicare, how="outer", on="FIPS") 
lifemedicare.head()

# remove weird columns at end of dataframe

lifemedicare = lifemedicare.drop(lifemedicare.index[3142:])

# read to csv
lifemedicare.to_csv('Output/lifemedicare.csv')

########### THE MASTER ALL DEATH CSV ##########

AllDeath = pd.merge(svi_merged, lifemedicare, how="outer", on="FIPS") 

AllDeath = AllDeath.drop(AllDeath.index[3142:])

AllDeath.to_csv('Output/AllDeath.csv')
```

```python
# read in death by gender data

genderdeaths = pd.read_csv('Resources/DeathsbyGender.txt', sep="\t")
```

```python
genderdeaths.head()
```

```python
# clean up columns

genderdeaths = genderdeaths.rename(columns={'ICD-10 113 Cause List': 'Cause of Death', 'Crude Rate': 'Rate per 1000'})
genderdeaths = genderdeaths[["Gender", "Cause of Death", "Deaths", "Population", "Rate per 1000"]]
```

```python
# replace medical terminology of ICD-10 cause list to layperson language

genderdeaths['Cause of Death'] = genderdeaths['Cause of Death'].replace('#Malignant neoplasms (C00-C97)', 'Cancer')

genderdeaths['Cause of Death'] = genderdeaths['Cause of Death'].replace('#Diseases of heart (I00-I09,I11,I13,I20-I51)', 'Heart Disease')

genderdeaths['Cause of Death'] = genderdeaths['Cause of Death'].replace('#Cerebrovascular diseases (I60-I69)', 'Stroke')

genderdeaths['Cause of Death'] = genderdeaths['Cause of Death'].replace('#Chronic lower respiratory diseases (J40-J47)', 'Respiratory')

genderdeaths['Cause of Death'] = genderdeaths['Cause of Death'].replace('#Accidents (unintentional injuries) (V01-X59,Y85-Y86)', 'Accidents')
```

```python
genderdeaths.head()
```

```python
# add overall death and death rates by gender to dataframe

genderdeaths = genderdeaths.append(pd.Series(['Male', 'Overall', 1400232, 159078923, 8.80212], index=genderdeaths.columns), ignore_index=True)
genderdeaths = genderdeaths.append(pd.Series(['Female', 'Overall', 1344016, 164048590, 8.19279], index=genderdeaths.columns), ignore_index=True)
```

```python
# add column of perecent of death based on conditional formatting

genderdeaths.loc[genderdeaths['Gender'] == 'Female', 'Percent'] = (genderdeaths['Rate per 1000'] / 8.19279) * 100
genderdeaths.loc[genderdeaths['Gender'] == 'Male', 'Percent'] = (genderdeaths['Rate per 1000'] / 8.80212) * 100
```

```python
# save to csv file 

genderdeaths.to_csv('Output/percentgenderdeath.csv')
```

```python
# read in deaths by density clean up columns like with gender deaths

densedeaths = pd.read_csv('Resources/UrbanRuralDeaths100k.txt', sep="\t")

densedeaths = densedeaths.rename(columns={'ICD-10 113 Cause List': 'Cause of Death', 'Crude Rate': 'Rate per 100k',
                                           '2013 Urbanization': 'Density'})
densedeaths = densedeaths[["Density", "Cause of Death", "Deaths", "Population", "Rate per 100k"]]

densedeaths['Cause of Death'] = densedeaths['Cause of Death'].replace('#Malignant neoplasms (C00-C97)', 'Cancer')

densedeaths['Cause of Death'] = densedeaths['Cause of Death'].replace('#Diseases of heart (I00-I09,I11,I13,I20-I51)', 'Heart Disease')

densedeaths['Cause of Death'] = densedeaths['Cause of Death'].replace('#Cerebrovascular diseases (I60-I69)', 'Stroke')

densedeaths['Cause of Death'] = densedeaths['Cause of Death'].replace('#Chronic lower respiratory diseases (J40-J47)', 'Respiratory')

densedeaths['Cause of Death'] = densedeaths['Cause of Death'].replace('#Accidents (unintentional injuries) (V01-X59,Y85-Y86)', 'Accidents')

densedeaths['Cause of Death'] = densedeaths['Cause of Death'].replace('#Intentional self-harm (suicide) (*U03,X60-X84,Y87.0)', 'Suicides')
```

```python
densedeaths.head()
```

```python
# read in txt file of total deaths by density and clean it up

test = pd.read_csv('Resources/total.txt', sep="\t")
test = test.rename(columns={'Notes': 'Cause of Death', 'Crude Rate': 'Rate per 100k', '2013 Urbanization': 'Density'})
test = test[["Density", "Cause of Death", "Deaths", "Population", "Rate per 100k"]]
```

```python
# append total deaths to dense deaths dataframe

densedeaths = densedeaths.append(test, ignore_index=True)
densedeaths = densedeaths.fillna('Total', downcast='infer')
```

```python
# add percentage by death rate column

densedeaths.loc[densedeaths['Density'] == 'Large Central Metro', 'Percent'] = (densedeaths['Rate per 100k'] / 716.52057) * 100
densedeaths.loc[densedeaths['Density'] == 'Large Fringe Metro', 'Percent'] = (densedeaths['Rate per 100k'] / 786.57736) * 100
densedeaths.loc[densedeaths['Density'] == 'Medium Metro', 'Percent'] = (densedeaths['Rate per 100k'] / 884.81205) * 100
densedeaths.loc[densedeaths['Density'] == 'Small Metro', 'Percent'] = (densedeaths['Rate per 100k'] / 962.16813) * 100
densedeaths.loc[densedeaths['Density'] == 'Micropolitan (Nonmetro)', 'Percent'] = (densedeaths['Rate per 100k'] / 1065.98419) * 100
densedeaths.loc[densedeaths['Density'] == 'NonCore (Nonmetro)', 'Percent'] = (densedeaths['Rate per 100k'] / 1200.95424) * 100
```

```python
# dense deaths to csv

densedeaths.to_csv('Output/densitydeath.csv')
```
